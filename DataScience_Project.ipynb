Data Visualization

```python
#scatterplot
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 3))
sns.scatterplot(x='Disease',y='Age',data=df)
plt.xlabel(' Disease')
plt.ylabel(' Age')
plt.title('Scatter Plot: Disease vs Age')
plt.show()
```

**Observation(Disease and Age):** The scatter plot aims to visualize the relationship between the 'Disease' and 'Age' columns in your DataFrame (df).

**Target Variable:** It's likely that 'Disease' is a categorical variable representing different types of diseases, and 'Age' is a numerical variable representing the age of patients or individuals.

**Patterns (Possible):** Looks for clusters or trends showing if specific diseases are more common at certain ages.

```python
#box plot
plt.figure(figsize=(6, 3))
sns.boxplot(x='Gender',y='Age',data=df,color='lightcoral')
plt.xlabel(' Gender')
plt.ylabel(' Age')
plt.title('Box Plot: Gender vs Age')
plt.show()
```

**Observation:** This plot compares the age distribution between genders using boxes and whiskers.

**Target:** We want to check for age differences between males and females.

**Pattern:** Look for variations in the box's position, size, and whiskers for each gender:

*  Median (line inside the box): If the median line is higher for one gender, it suggests that that gender tends to have higher ages.

*  Box size: A larger box indicates more age variability for that gender.


*  Whiskers: They show the range of ages.Longer whiskers suggest a wider age range.

*  Outliers (dots outside whiskers): Represent unusual age values for that gender.

```python
#bar chart
plt.figure(figsize=(6, 3))
sns.barplot(x='Blood Pressure',y='Age',data=df,color='palegreen')
plt.xlabel(' Blood Pressure')
plt.ylabel(' Age')
plt.title('Bar Chart: Blood Pressure vs Age')
```

**Observation:** This plot shows how blood pressure and age relate. Each bar represents a blood pressure category, and its height corresponds to the average age of people within that category.

**Target:** We aim to see if there's a relationship between blood pressure and age—specifically, if blood pressure tends to change as people get older.

**Pattern:** Look for trends in the heights of the bars.If the bars generally get taller or shorter as you move along the x-axis (blood pressure categories), it suggests a trend. For example, if the bars get taller as blood pressure increases, it could indicate that older people tend to have higher blood pressure.

```python
#histogram
plt.figure(figsize=(6, 4))
sns.histplot(df['Disease'],color='orange',kde=True)
plt.xlabel(' Disease')
plt.ylabel(' Frequency')
plt.title('Histogram: Disease')
plt.show()
```

**Observation:** This plot shows the frequency of each disease in your dataset. Think of it like counting how many times each disease appears.

**Target:** The goal is to understand which diseases are more or less common in your data.

**Pattern:** Look at the height of each bar in the histogram.

* Taller bars: Indicate diseases that occur more frequently. These are the common diseases in your dataset.
* Shorter bars: Indicate diseases that occur less frequently. These are less common in your data.

Data encoding

Model training

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
df=df.dropna()
le_Outcome_Variable=LabelEncoder()
#target encoding
df['Outcome Variable']=le_Outcome_Variable.fit_transform(df['Outcome Variable'])
#other categorical features encoding
df['Gender']=LabelEncoder().fit_transform(df['Gender'])
df['Disease']=LabelEncoder().fit_transform(df['Disease'])
df['Cough']=LabelEncoder().fit_transform(df['Cough'])
df['Fatigue']=LabelEncoder().fit_transform(df['Fatigue'])
df['Fever']=LabelEncoder().fit_transform(df['Fever'])
df['Blood Pressure']=LabelEncoder().fit_transform(df['Blood Pressure'])
df['Cholesterol Level']=LabelEncoder().fit_transform(df['Cholesterol Level'])
df['Difficulty Breathing']=LabelEncoder().fit_transform(df['Difficulty Breathing'])
```

```python
#features selection
x=df.drop('Outcome Variable',axis=1)
y=df['Outcome Variable']
#train, test spliting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, recall_score, precision_score,root_mean_squared_error
#train model
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
#predict
y_pred=model.predict(x_test)
#evaluate
print(f"Accuracy: {accuracy_score(y_test,y_pred)}")
print(f"Classification Report:\n{classification_report(y_test,y_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,y_pred)}")
print(f"Recall: {recall_score(y_test, y_pred, average='weighted')}")
print(f"Precision: {precision_score(y_test, y_pred, average='weighted')}")
print(f"Root Mean Squared Error: {root_mean_squared_error(y_test, y_pred)}")
```

```python
#heat map Confusion Matrix
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(6, 3))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues',
            xticklabels=LabelEncoder().fit_transform(df['Outcome Variable'].unique()),
            yticklabels=LabelEncoder().fit_transform(df['Outcome Variable'].unique()))
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
